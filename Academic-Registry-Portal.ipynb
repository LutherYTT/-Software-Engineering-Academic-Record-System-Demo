```python
def move_center(window):
    screen_width, screen_height = window.get_screen_dimensions()
    win_width, win_height = window.size
    x, y = (screen_width - win_width)//2, (screen_height - win_height)//2
    window.move(x, y)
    logging.info('Move window to center.')
```

```python
#check is the string have any number
def contains_number(string):
    return any(char.isdigit() for char in string)
```

```python
#ITT version
import pandas as pd
import PySimpleGUI as sg

#import data, locate csv file here
df = pd.read_csv("student-dataset.csv")

selected_row = 0

#setting data
df_table = pd.DataFrame(df)
data_headings = list(df.columns[:])
data_values = df_table.values.tolist()
#set the ITT_windows performance
sg.theme("DarkBlue3")
sg.set_options(font=("Courier New", 16))
sg.set_options(tooltip_font=("Courier New", 13))
Leave = False

ITT_title = [sg.Text('IT Technician Version',justification='center')]
ITT_table = [[sg.Table(values=data_values, headings=data_headings, 
                    auto_size_columns=True, justification='center',
                    enable_events=True, key='-filestable-',vertical_scroll_only = False,
                    enable_click_events=True, display_row_numbers=True, size = (10,10))]]
                    #col_widths=list(map(lambda x:len(x)+3, data_headings
                    #)))]]
ITT_search = [[sg.Text('Search value: '), sg.InputText(key='search_input', size=(15, 10)), 
           sg.Button('Search',key='-search-'), sg.Button('clear',key='-clear-')]]

logging.info("ITT version setup...")

#Update the GUI after click update button
def ITT_update_layout_record(df):
    global order, data_values, df_table, ITT_window
    
    logging.info('User end the input process.')
    
    ITT_title = [sg.Text('IT Technician Version',justification='center')]
    ITT_table = [[sg.Table(values=data_values, headings=data_headings, 
                    auto_size_columns=True, justification='center',
                    enable_events=True, key='-filestable-', vertical_scroll_only = False,
                    enable_click_events=True, display_row_numbers=True)]]
                    #col_widths=list(map(lambda x:len(x)+3, data_headings
                    #)))]]
    ITT_search = [[sg.Text('Search value: '), sg.InputText(key='search_input', size=(15, 10)), 
               sg.Button('Search',key='-search-'), sg.Button('clear',key='-clear-')]]
    ITT_temp_layout = [[ITT_title],[sg.Push(), ITT_table, sg.Push()],
                      [ITT_search],[sg.Button('Logout',key='-logout-')]]
    #return the table become original GUI
    return(ITT_temp_layout)

#Update GUI and provide the input box to edit record
def ITT_update_layout_input(data_selected, row, df):
    global order, data_values, df_table, ITT_window
    
    logging.info('User start to input...')
    
    ITT_title = [sg.Text('IT Technician Version',justification='center')]
    ITT_table = [[sg.Table(values=data_values, headings=data_headings, 
                    auto_size_columns=True, justification='center',
                    enable_events=True, key='-filestable-', vertical_scroll_only = False,
                    enable_click_events=True, display_row_numbers=True)]]
                    #col_widths=list(map(lambda x:len(x), data_headings
                    #)))]]
    #the values of selected row
    selected_Studid = list(df.iloc[row][df.columns[0]])[0]
    selected_Name = list(df.iloc[row][df.columns[1]])[0]
    selected_Nation = list(df.iloc[row][df.columns[2]])[0]
    selected_Gender = list(df.iloc[row][df.columns[3]])[0]
    selected_Age = list(df.iloc[row][df.columns[4]])[0]
    selected_EngG = list(df.iloc[row][df.columns[5]])[0]
    selected_MathG = list(df.iloc[row][df.columns[6]])[0]
    selected_SciG = list(df.iloc[row][df.columns[7]])[0]
    selected_LangG = list(df.iloc[row][df.columns[8]])[0]
    selected_PortR = list(df.iloc[row][df.columns[9]])[0] 
    selected_CovR = list(df.iloc[row][df.columns[10]])[0] 
    selected_RefR = list(df.iloc[row][df.columns[11]])[0] 
    #input box performance
    input_box = [[sg.InputText(selected_Studid, size=(8, 10), key='-col0-', tooltip="Student ID"), #input box ofstudent ID
            sg.InputText(selected_Name, size=(10, 10), key='-col1-', tooltip="Name"), #input box of Name
            sg.InputText(selected_Nation, size=(10, 10), key='-col2-', tooltip="Nationality"), #Input box of Nationality
            sg.InputText(selected_Gender, size=(6, 10), key='-col3-', tooltip="Gender"), #Input box of Gender
            sg.InputText(selected_Age, size=(5, 10), key='-col4-', tooltip="Age"), #Input box of Gender
            sg.InputText(selected_EngG, size=(4, 10), key='-col5-', tooltip=df.columns[5]), #Input box of English Grade
            sg.InputText(selected_MathG, size=(4, 10), key='-col6-', tooltip=df.columns[6]), #Input box of Math Grqade
            sg.InputText(selected_SciG, size=(4, 10),  key='-col7-', tooltip=df.columns[7]), #Input box of Science Grade
            sg.InputText(selected_LangG, size=(4, 10), key='-col8-', tooltip=df.columns[8]), #Input box of Language Grade  
            sg.InputText(selected_PortR, size=(4, 10),key='-col9-', tooltip=df.columns[9]), #Input box of Portfolio Rating
            sg.InputText(selected_CovR, size=(4, 10), key='-col10-', tooltip=df.columns[10]), #Input box of Coverletter Rating
            sg.InputText(selected_RefR, size=(4, 10), key='-col11-', tooltip=df.columns[11]), #Input box of Refletter Rating
                 ]] 
        
    ITT_layout = [[ITT_title],
                [sg.Push(), ITT_table, sg.Push()],
                [sg.Push(), input_box, sg.Push()],
                [sg.Button('Update', key='-update-'),sg.Button('Back', key='-back-'), sg.Button('Exit'), sg.Button('Logout',key='-logout-')]]
    #ITT_window update
    #return the table which have the input box
    return(ITT_layout)

#------------------------------------------#
def ITT_check_input(input_box_record):
    logging.info('Checking the data editing in put format...')
    
    special_characters1 = "!@#$%^&*()-+?_=.,<>'/"
    special_characters2 = '"'
    studID = input_box_record[0]
    Gender = input_box_record[3]
    Name = input_box_record[1]
    Nation = input_box_record[2]
    age = input_box_record[4]
    grade_list=input_box_record[5:]
        
    if(len(studID) != 8): 
        return('studID ERROR! \n > Student ID should have 8 digit !')
    elif(studID[0:1] != 's'): # format require
        return('studID ERROR! \n > Student ID should start with s !')
    elif(not(studID[1:].isdigit())): # last 8 digit number only
        return('studID ERROR! \n > Student ID last 7 digit should be number !')
    # no special character
    elif any((sc in special_characters1 for sc in studID) or (sc in special_characters2 for sc in studID)):
        return('studID ERROR! \n > Name should not contains any special character !')
    
    elif(not(Gender == 'Male' or Gender == 'Female')):
        return('Gender ERROR! \n > Gender should be "Male" or "Female" !')
    
     # no number in Name
    elif(contains_number(Name)):
        return('Name ERROR! \n > Name should not contains number !')
    # no special character
    elif any((sc in special_characters1 for sc in Name)or(sc in special_characters2 for sc in Name)):
        return('Name ERROR! \n > Name should not contains any special character !')
    
    # no number in Nationality
    elif(contains_number(Nation)):
        return('Nationality ERROR! \n > Nationality should not contains number !')
    # no special character
    elif any((sc in special_characters1 for sc in Nation)or(sc in special_characters2 for sc in Nation)):
        return('Nationality ERROR! \n > Nationality should not contains any special character !')
    
    # no number in Age
    elif(not(age[:].isdigit())): # last 8 digit number only
        return('Age ERROR! \n > Age should be number !')
    # no special character
    elif any((sc in special_characters1 for sc in age)or(sc in special_characters2 for sc in age)):
        return('Age ERROR! \n > Age should not contains any special character !')
    
    for i in range(0,(len(grade_list))):
        if(not(contains_number(grade_list[i]))): # Grade number only
            return('Grade ERROR! \n > Grade should be number !')
        elif(float(grade_list[i]) > 5.0 or float(grade_list[i]) < 0.0):
            return('Grade ERROR! \n > Grade and Rating should be number and between 0 and 5 !')
    
#---------------(Sort the data)---------------#
order = 1
def ITT_sort_col(temp_row, temp_col):
    global order, data_values, df_table, ITT_window, ITT_table
    temp = df.columns[temp_col]
    temp_order = False
    if(order%2 ==0): 
        temp_order = True
    df_table = df_table.sort_values(by=[temp], ascending=temp_order)
    data_values = df_table.values.tolist()
    order += 1
    #update the table be the sorted version
    ITT_window['-filestable-'].update(values=data_values)
    
    logging.info('Table got sorted.')
    
    return(df_table)

#---------------(Record GUI)---------------#
def ITT_show_table():
    global order, data_values, df_table, ITT_window, ITT_table, Leave
    now_df = df
    #setting the table performance
    ITT_layout = [[ITT_title],[sg.Push(), ITT_table, sg.Push()],[ITT_search],[sg.Button('Logout',key='-logout-')]]
    #print ITT_window
    ITT_window = sg.Window('Academic record System(ITT)', ITT_layout, finalize = True, margins=(100, 100))
    move_center(ITT_window)
    while True:
        event, values = ITT_window.read()
        if event in (sg.WIN_CLOSED, 'Exit'):
            ITT_window.close()
            Leave = True
            logging.info('System close.')
            break
            
        elif event == '-back-':
            #update the dataframe
            now_df = df
            df_table = pd.DataFrame(now_df)
            data_values = df_table.values.tolist()
            #update the GUI
            temp = ITT_update_layout_record(now_df)
            ITT_window.close()
            ITT_window = sg.Window('Academic record System(ITT)', temp, finalize = True, margins=(100, 100))
            move_center(ITT_window)

        #When user selected the row, show the input box
        elif event == '-filestable-': 
            data_selected = [data_values[row] for row in values[event]]
            selected_row = values[event]
            #update the GUI
            ITT_update_layout = ITT_update_layout_input(data_selected, values[event], now_df)
            ITT_window.close()
            ITT_window = sg.Window('Academic record System(ITT) - INPUT', ITT_update_layout, finalize = True, margins=(100, 100))
            move_center(ITT_window)
            
            logging.info('Data got selected.')
            
        #when user click the update button
        elif event == '-update-':
            input_box_record = [values['-col0-'],values['-col1-'],values['-col2-'],values['-col3-'],values['-col4-'],
                                values['-col5-'],values['-col6-'],values['-col7-'],values['-col8-'],values['-col9-'],
                                values['-col10-'],values['-col11-']]
            temp = ITT_check_input(input_box_record)
            #if the check_input return the warning message, than pop the small GUI to warning
            if isinstance(temp, str):
                sg.popup(temp)
                logging.warning('Input format error, fail the update. Reason: {}'.format(temp))
            else:
                #if not message, than update the record
                now_df.iat[selected_row[0], 0] = input_box_record[0]  #StudID
                now_df.iat[selected_row[0], 1] = input_box_record[1]  #Name
                now_df.iat[selected_row[0], 2] = input_box_record[2]  #Nationality
                now_df.iat[selected_row[0], 3] = input_box_record[3]  #Gender
                now_df.iat[selected_row[0], 4] = input_box_record[4]  #Age
                now_df.iat[selected_row[0], 5] = float(input_box_record[5])  #EngG
                now_df.iat[selected_row[0], 6] = float(input_box_record[6])  #MathG
                now_df.iat[selected_row[0], 7] = float(input_box_record[7])  #SciG
                now_df.iat[selected_row[0], 8] = float(input_box_record[8])  #LangG
                now_df.iat[selected_row[0], 9] = float(input_box_record[9])  #PortR
                now_df.iat[selected_row[0], 10] = float(input_box_record[10])  #CovR
                now_df.iat[selected_row[0], 11] = float(input_box_record[11]) #RefR
                logging.info('Update Success.')
            #update the table become edited table
            df_table = pd.DataFrame(now_df)
            data_values = df_table.values.tolist()
            #get the GUI layout of original GUI
            temp = ITT_update_layout_record(now_df)
            ITT_window.close()
            ITT_window = sg.Window('Academic record System(ITT)', temp, finalize = True, margins=(100, 100))
            move_center(ITT_window)
            pass
        elif event == '-search-':
            search_value = values['search_input']
            # 遍历表格数据，查找特定值
            df_search = df[df.apply(lambda row: row.astype(str).str.contains(search_value, case=True).any(), axis=1)]
            now_df = df_search
            data_search_result = df_search.values.tolist()
            ITT_window['-filestable-'].update(values=data_search_result)
            logging.info('User searching: {}'.format(search_value))

        elif event == '-clear-':
            #clear the search input box, and make the GUI become original GUI
            df_search = df[df.apply(lambda row: row.astype(str).str.contains('', case=True).any(), axis=1)]
            now_df = df_search
            data_values = df_search.values.tolist()
            values['search_input'] = ''
            ITT_window['search_input'].update(value='')
            ITT_window['-filestable-'].update(values=data_values)
            logging.info('User clear the search')

        elif event == '-logout-':
            ITT_window.close()
            Leave = False
            logging.info('User logot.')
            break
        #when click the header
        elif '+CLICKED+' in event:
            temp_row = event[2][0]
            temp_col = event[2][1]
            if(temp_row == -1 and temp_col != -1):
                now_df = ITT_sort_col(temp_row, temp_col)
    if(not(Leave)):
        login()
```

```python
#Login System
import PySimpleGUI as sg
import pandas as pd
df = pd.read_csv("student-dataset.csv")
temp = df
sg.theme("DarkBlue3")
sg.set_options(font=("Courier New", 16))

#---------------(Check the Account exist or not)---------------#
def validation(temp_acc, temp_pw, temp_role, Login_window):
    if(temp_role == 'Student'):
        df_DataMatch = df[df.apply(lambda row: row.astype(str).str.contains(temp_acc, case=True).any(), axis=1)]
        try:
            password = df_DataMatch.iloc[0, 0][1:]
        except:
            return False
        if(temp_pw == password):
            logging.info('Student login.')
            Login_window.close()
            return('S')
        else: return False
        pass
    elif(temp_role == 'Teacher'):
        login_data = {'acc' : "s123456", 'pw' : 't123'},{'acc' : 's56789', 'pw' : 't567'}
        for temp in login_data:
            if temp['acc'] == temp_acc:
                if temp_pw == temp['pw']:
                    logging.info('Teacher login.')
                    Login_window.close()
                    return('T')
                    break
            else: return False
        pass
    elif(temp_role == 'IT Technician'):
        login_data = {'acc' : "s123456", 'pw' : 'itt123'},{'acc' : 's56789', 'pw' : 'itt567'}
        for temp in login_data:
            if temp['acc'] == temp_acc:
                if temp_pw == temp['pw']:
                    logging.info('IT Technician login.')
                    Login_window.close()
                    return('ITT')
                    break
            else: return False
        pass
    pass

#---------------(Check login input format)---------------#
def check_login_input(temp_acc, temp_pw):
    special_characters1 = "!@#$%^&*()-+?_=.,<>'/"
    special_characters2 = '"'
    input_length_limit = 16

    if(len(temp_acc) > input_length_limit): # too length
        return('Account ERROR! \n > Account should in 16 digit !')
    elif(temp_acc[0:1] != 's'): # format require
        return('Account ERROR! \n > Account should start with "s" !')
    elif(not(temp_acc[1:].isdigit())):
        return('Account ERROR! \n > Without first character, other digit must be number !')
    elif any((sc in special_characters1 for sc in temp_acc) or (sc in special_characters2 for sc in temp_acc)):
        return('Account ERROR! \n > Account should not contains any special character !')
    
    elif(len(temp_pw) > input_length_limit): # too length
        return('Password ERROR! \n > Account should in {} digit !'.format(input_length_limit))
    elif any( (sc in special_characters1 for sc in temp_pw)or(sc in special_characters2 for sc in temp_pw)):
        return('Password ERROR! \n > Password should not contains any special character !')
    
#---------------(Login GUI)---------------#
def login():
    count = 0
    role = ['Student', 'Teacher', 'IT Technician']
    but_color = ['white on green', 'white on red', 'white on blue']
    
    title_message =  [[sg.Text('Welcome to ',justification='center')],
              [sg.Text('School of Science and Technology',justification='center')], 
              [sg.Text('Academic record System',justification='center')]]
                      
    Login_layout = [ [sg.Push(), sg.Column(title_message, element_justification='center'), sg.Push()],   
             [sg.Text('Enter Account\t'), sg.InputText()],
             [sg.Text('Enter Password\t'), sg.InputText(password_char='*')],
            # choose the user role
             [sg.Button(role[count], size=(15,1), button_color=but_color[count], key='-Role-')],  
             [sg.Button('Login', key='-Login-', size=(5,1)), sg.Button('Exit', size=(5,1))]]

    Login_window = sg.Window('Login UI Testing', Login_layout, margins=(50, 50), grab_anywhere=True)
    
    while True:             # Event Loop
        event, values = Login_window.read()
        logging.info('User trying to login...')
        print(event, values)
        if event in (sg.WIN_CLOSED, 'Exit'):
            Login_window.close()
            break
        elif event == '-Login-':
            input_acc = values[0]
            input_pw = values[1]
            print('{},{},{}'.format(values[0], values[1], role[count]))
            temp = check_login_input(input_acc, input_pw)
            if isinstance(temp, str):
                sg.popup(temp)
                logging.warning('User input format error, fail the login.')
            elif (validation(input_acc, input_pw, role[count], Login_window) == False):
                sg.popup('Wrong account or Password !')
                logging.warning('User input data error, fail the login.')
            elif(validation(input_acc, input_pw, role[count], Login_window) in ['S','T','ITT']):
                login_version = validation(input_acc, input_pw, role[count], Login_window)
                break
        elif event == '-Role-':                # if role button clicked, change role
            count += 1
            if count>=3: count = 0
            Login_window['-Role-'].update(text=role[count], button_color=but_color[count])
        
    if(login_version == 'S'): stud_show_table(input_acc)
    elif(login_version == 'T'): show_table()
    elif(login_version == 'ITT'): ITT_show_table()
```
